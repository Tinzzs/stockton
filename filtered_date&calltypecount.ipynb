```python
df["CALL_ENTRY_DATE_F"] = pd.to_datetime(df["CALL_ENTRY_DATE"], format='mixed')
```

```python
mask = (df["CALL_ENTRY_DATE_F"] >= "2019-01-01") & (df["CALL_ENTRY_DATE_F"] <= "2023-12-31")
filtered_df = df.loc[mask]
```

```python
# filtered_df.drop(columns=['CALL_ENTRY_DATE'], inplace=True)
filtered_df.head(-10)
```

```python
filtered_df.to_csv("/mnt/c/Users/huyti/OneDrive/Documents/anaconda_projects/CIVL293/Stockton/Data/stockton_911calls_25.csv", index=False, encoding="utf-8-sig")
```

```python
#2025 data used for benchmarking final models  
#Enfore consistent type for mismatched 911 call type count
filtered_df["CALL_TYPE_ORIG"] = filtered_df["CALL_TYPE_ORIG"].astype(str).str.strip().str.upper()
filtered_df["CALL_TYPE_FINAL"] = filtered_df["CALL_TYPE_FINAL"].astype(str).str.strip().str.upper()

mismatch_per = (filtered_df["CALL_TYPE_ORIG"] != filtered_df["CALL_TYPE_FINAL"]).mean()*100
mismatch = (filtered_df["CALL_TYPE_ORIG"] != filtered_df["CALL_TYPE_FINAL"]).sum()
match = (filtered_df["CALL_TYPE_ORIG"] == filtered_df["CALL_TYPE_FINAL"]).sum()
diffna = (filtered_df["CALL_TYPE_ORIG"].fillna("NA") != filtered_df["CALL_TYPE_FINAL"].fillna("NA")).sum()
print("Mismatch counts: ", mismatch)
print("Percent mismatch", mismatch_per)
print("Matches", match)
total = len(filtered_df)
print("Mismatch counts with filled NA", diffna)
unc = total - (match + mismatch)
print("Unaccounted rows:", unc)
```

```python
filtered_df.shape
```

```python
# Top mismatched pairs of 2025
mismatches_df = filtered_df[filtered_df["CALL_TYPE_ORIG"] != filtered_df["CALL_TYPE_FINAL"]]
top_pairs = mismatches_df.groupby(["CALL_TYPE_ORIG", "CALL_TYPE_FINAL"]).size().sort_values(ascending=False).head(10)
print("\nTop mismatching pairs:\n", top_pairs)
```

```python
# Top mismatched crime type derived from 911 call type of 2025
df3 = pd.read_csv("/mnt/c/Users/huyti/OneDrive/Documents/anaconda_projects/CIVL293/Stockton/Data/output_25.csv")
mismatches_df = df3[df3["CALL_TYPE_ORIG"] != df3["CALL_TYPE_FINAL"]]
mismatch = (df3["CALL_TYPE_ORIG"] != df3["CALL_TYPE_FINAL"]).sum()

category_counts = mismatches_df["category"].value_counts()
print("Total:", len(df3), "\n")
print("Total mismatches", mismatch, "\n")
print("Top categories contributing to mismatches:\n", category_counts)
```
